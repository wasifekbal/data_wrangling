```python
# Creating copies

patients_copy=patients.copy()
treatment_copy=treatment.copy()
adverse_copy=adverse.copy()
```

## A. Completeness Issues

- Missing values for contact details of 12 patients
- Only 280 rows are there, instead of 350
- hba1c_change col has missing values

#### Define and Code

```python
# Removing 12 patients whose contact details can not be found.
patients_copy.dropna(inplace=True)
```

```python
# Appending DataFrame treatment_cut in treatment_copy to fullfill 70 missing patients.
treatment_copy=treatment_copy.append(tcut,ignore_index=True)
del(tcut)
```

```python
# Subtract hba1c_start from hba1c_end to get hba1c_change
treatment_copy.drop(columns={'hba1c_change'},inplace=True)
treatment_copy['hba1c_change']=treatment_copy['hba1c_start']-treatment_copy['hba1c_end']
```

## B. Tidiness Issues

- Phone + email in contact col `patients`
- Novodra and Aurolin col `treatment`
- Merge the `adverse` table to `treatment` table

### Define and Code

```python
# Novodra and Aurolin using the melt function
treatment_copy=pd.melt(treatment_copy, id_vars=['given_name', 'surname', 'hba1c_start', 'hba1c_end', 'hba1c_change'],
                           var_name='treatment', value_name='dose')

treatment_copy=treatment_copy[treatment_copy['dose']!='-']

treatment_copy['start'],treatment_copy['end']=treatment_copy['dose'].str.split('-').str

treatment_copy.drop(columns={'dose'}, inplace=True)

treatment_copy['start']=treatment_copy['start'].str.replace('u','')

treatment_copy['end']=treatment_copy['end'].str.replace('u','')

treatment_copy['start']=treatment_copy['start'].astype('int')

treatment_copy['end']=treatment_copy['end'].astype('int')
```

```python
# Merging the `adverse` table to `treatment` table using pd.merge() function.
# as information of advarse reaction doesn't need a separate table.

treatment_copy=treatment_copy.merge(adverse_copy,on="given_name",how="left")

treatment_copy.drop(columns={'surname_y'},inplace=True)

treatment_copy.rename(columns={'given_name':'fname','surname_x':'lname'},inplace=True)

del(adverse_copy)

del(adverse)
```

```python
# Extracting the phone number and email address from contact column using re (regular expression).

phone_numbers=[]
emails=[]
for ph in patients_copy['contact']:
    try:
        match=re.findall(r'\d{3}-\d{3}-\d{4}',ph)[0]
        phone_numbers.append(match)
    except:
        try:
            match=re.findall(r'\d{3} \d{3} \d{4}',ph)[0]
            match='-'.join(match.split())
            phone_numbers.append(match)
        except:
            try:
                match=re.findall(r'\(\d{3}\) \d{3}-\d{4}',ph)[0]
                match1=re.findall(r'\(\d{3}\)',match)[0][1:-1]
                match2=re.findall(r'\d{3}-\d{4}',match)[0]
                
                match=match1+'-'+match2
                phone_numbers.append(match)
            except:
                phone_numbers.append(np.nan)
    try:
        match=re.findall(r'[a-zA-z]\w{3,50}@[a-z]{2,20}.[a-z]{2,3}',ph)[0]
        emails.append(match)
    except:
        emails.append(np.nan)
```

```python
# Inserting the extracted phone number and email into the treatment_copy dataframe and removing the original 'contact' column.

patients_copy['email']=emails

patients_copy['phone_number']=phone_numbers

patients_copy.drop(columns='contact',inplace=True)
```

- Zip code is in float format.
- some zip codes are less than 5 digits ( it is inconsistency as well as validity issue. )

### Code 

```python
patients_copy['zip_code']=patients_copy['zip_code'].astype('int')

zipcode=[]
for i in patients_copy['zip_code']:
    x=list(str(j) for j in str(i))
    if len(x)<5:
        x.insert(0,'0')
    x=''.join(x)
    zipcode.append(x)
    
patients_copy['zip_code']=zipcode
```

### Assessing Data

In this step, the data is to be understood more deeply. Before implementing methods to clean it, you will definitely need to have a better idea about what the data is about.

### Types of Unclean Data

There are 2 kinds of unclean data

1. <u>Dirty Data</u> (Data with Quality issues):<br>
    Dirty data, also known as low quality data. Low quality data has content issues.<br>
2. <u>Messy Data</u> (Data with tidiness issues)<br>
    Messy data, also known as untidy data. Untidy data has structural issues.

### 1. Dirty Data

When the data has following kinds of issues:

1. Duplicated data
2. Missing Data
3. Corrupt Data
4. Inaccurate Data

This kind of data is known as Dirty data

### 2. Messy Data

When the data has structural issue is it known as Messy or untidy data.

Tidy data has the following properties:

1. Each variable forms a column
2. Each observation forms a row
3. Each observational unit forms a table

If a data violates any one of the above it is known as untidy data
